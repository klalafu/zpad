```python
import pandas as pd
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
```

```python
df = pd.read_csv("/home/kali/Downloads/kidney_disease.csv", sep=",")
```

1. Боротьба зі зниклими даними

```python
df.head()
```

```python
df = df.drop("id", axis=1)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['age']=df['age'].fillna(df['age'].mean())

print("Rows before droping: ", df.shape[0])
df.dropna(axis=0, thresh=20,inplace=True)
print("Rows after droping: ", df.shape[0])

print("Columns before droping: ", df.shape[1])
df.dropna(axis=1, thresh=310,inplace=True)
print("Columns after droping: ", df.shape[1])

df['sg']=df['sg'].fillna(df['sg'].mean())
```

```python
df = df.bfill()
df.reset_index(drop=True, inplace=True)
df.info()
```

```python
df.loc[53, "pcv"] = df.loc[52, "pcv"] 
df['pcv'] = pd.to_numeric(df['pcv'])
```

```python
df.info()
```

```python
for column in df.select_dtypes('object').columns:
    print(column, df[column].unique())
```

```python
df['dm'] = df['dm'].replace(to_replace = '\tyes', value = 'yes') 
df['cad'] = df['cad'].replace(to_replace = '\tno', value = 'no') 
df['classification'] = df['classification'].replace(to_replace = 'ckd\t', value = 'ckd') 
```

```python
for column in df.select_dtypes('object').columns:
    print(column, df[column].unique())
```

2. Нормалізація

```python
def normalization(df, column):
    df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())

df_norm = df.copy()
num = ["float64", "int64"]
for column in df.select_dtypes(include=num).columns:
    if column == 'age': continue
    normalization(df_norm, column)
df_norm.head()
```

3. Збудувати гістограму по одному із атрибутів, що буде показувати на кількість елементів, що знаходяться у 10 діапазонах, які ви задасте.

```python
bin_edges = [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6]

plt.figure(figsize=(10, 6))
sns.histplot(df_norm['pcv'], bins=bin_edges, kde=False, edgecolor='black')
plt.xlabel('pcv')
plt.show()
```

4. Збудувати графік залежності одного integer/real атрибута від іншого.

```python
plt.figure(figsize=(10, 6))
plt.xlabel('Packed cell volume')
plt.ylabel('Hemoglobin (g/dl)')
sns.lineplot(x=df_norm['pcv'], y=df_norm['hemo'], errorbar=None)
plt.show()
```

5. Підрахувати коефіцієнт Пірсона та Спірмена для двох integer/real атрибутів.

```python
scipy.stats.pearsonr(x=df_norm['pcv'], y=df_norm['hemo'])
```

```python
scipy.stats.spearmanr(df_norm['pcv'], df_norm['hemo'])
```

6. Провести One Hot Encoding категоріального string атрибуту.

```python
df_norm['pc'].head()
```

```python
one_hot = pd.get_dummies(df_norm['pc'])
one_hot.columns = ['pc_abnormal', 'pc_normal']
df_norm = df_norm.drop('pc',axis = 1)
df_norm = df_norm.join(one_hot)
df_norm.head()
```
